```python

proteins = []
url = "https://www.fpbase.org/api/proteins/?parent_organism=6100&format=json"

while url:
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    if isinstance(data, list):
        proteins.extend(data)
        break
    else:
        proteins.extend(data.get("results", []))
        url = data.get("next")

# Remove channelrhodopsins
exclude_terms = ('channelrhodopsin', 'rcamp', 'gcamp2', 'cp-mkate', 'cegfp')

proteins = [
    p for p in proteins
    if not any(term in (p.get('name') or '').lower() for term in exclude_terms)
]

# Compute Stokes shift
for protein in proteins:
    protein['stokes_shift'] = None

    for state in protein.get("states", []):
        ex_max = state.get("ex_max")
        em_max = state.get("em_max")

        if ex_max is not None and em_max is not None:
            shift = em_max - ex_max
            state["stokes_shift"] = shift
            protein["stokes_shift"] = shift
            break  # First valid state only

# Example check: print names and stokes shift
for p in proteins:
    print(f"{p.get('name')}: Stokes shift = {p.get('stokes_shift')}")

```

```python
import requests
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras import layers, models

# ------------------------------------------------
# 1. Download Aequorea victoria proteins (parent_organism=6100)
# ------------------------------------------------

proteins = []
url = "https://www.fpbase.org/api/proteins/?parent_organism=6100&format=json"

while url:
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    if isinstance(data, list):
        proteins.extend(data)
        break
    else:
        proteins.extend(data.get("results", []))
        url = data.get("next")

# Optional: remove families you don't want
exclude_terms = ('channelrhodopsin', 'rcamp', 'gcamp2', 'cp-mkate', 'cegfp')
proteins = [
    p for p in proteins
    if not any(term in (p.get('name') or '').lower() for term in exclude_terms)
]

print(f"Total protein records retrieved (A. victoria filtered): {len(proteins)}")

# ------------------------------------------------
# 2. Extract sequences and brightness
# ------------------------------------------------

seqs = []
brightness_vals = []

for entry in proteins:
    seq = entry.get("seq")
    states = entry.get("states", [])
    brightness = None

    if states and isinstance(states, list):
        # Try brightness, fallback to rel_brightness if needed
        for st in states:
            b = st.get("brightness")
            if b is None:
                b = st.get("rel_brightness")
            if b is not None:
                brightness = b
                break

    if seq and brightness is not None:
        seqs.append(seq)
        brightness_vals.append(brightness)

print(f"Total usable sequences with brightness: {len(seqs)}")

# Bail out early if nothing usable
if len(seqs) == 0:
    raise RuntimeError("No sequences with brightness found. Check FPbase fields.")

# ------------------------------------------------
# 3. One-hot encode sequences (fixed length = 238)
# ------------------------------------------------

max_len = 238  # EGFP length (truncate or zero-pad others)
aa_alphabet = 'ACDEFGHIKLMNPQRSTVWY'
aa_index = {aa: i for i, aa in enumerate(aa_alphabet)}

def one_hot_encode(seq: str, max_len: int) -> np.ndarray:
    arr = np.zeros((max_len, len(aa_alphabet)), dtype=np.float32)
    for i, aa in enumerate(seq[:max_len]):
        if aa in aa_index:
            arr[i, aa_index[aa]] = 1.0
    return arr.flatten()

X = np.array([one_hot_encode(seq, max_len) for seq in seqs], dtype=np.float32)
y = np.array(brightness_vals, dtype=np.float32)

print("X shape:", X.shape)  # (N, 238*20)
print("y shape:", y.shape)

# ------------------------------------------------
# 4. Quick look at brightness distribution
# ------------------------------------------------

plt.figure(figsize=(6, 4))
plt.hist(y, bins=20)
plt.xlabel("Brightness")
plt.ylabel("Count")
plt.title("Brightness distribution (A. victoria set)")
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 5. Train/test split and scaling
# ------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)

# Standardize features (helps NN training even though it's one-hot)
scaler = StandardScaler(with_mean=True, with_std=True)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ------------------------------------------------
# 6. Define neural network regressor
# ------------------------------------------------

input_dim = X_train_scaled.shape[1]

model = models.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # brightness (regression)
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='mse',
    metrics=['mae']
)

model.summary()

# ------------------------------------------------
# 7. Train the model
# ------------------------------------------------

history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=80,
    batch_size=32,
    verbose=1
)

# ------------------------------------------------
# 8. Plot training curves (loss & MAE)
# ------------------------------------------------

history_dict = history.history

plt.figure(figsize=(6, 4))
plt.plot(history_dict['loss'], label='Train loss')
plt.plot(history_dict['val_loss'], label='Val loss')
plt.xlabel("Epoch")
plt.ylabel("MSE loss")
plt.title("Training vs Validation Loss (brightness)")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(history_dict['mae'], label='Train MAE')
plt.plot(history_dict['val_mae'], label='Val MAE')
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.title("Training vs Validation MAE (brightness)")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 9. Evaluate on held-out test data
# ------------------------------------------------

test_loss, test_mae = model.evaluate(X_test_scaled, y_test, verbose=0)
print("Neural Network (brightness)")
print(f"  Test MSE: {test_loss:.3f}")
print(f"  Test MAE: {test_mae:.3f}")

# ------------------------------------------------
# 10. Predicted vs actual brightness scatter
# ------------------------------------------------

y_pred = model.predict(X_test_scaled, verbose=0).reshape(-1)

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--')
plt.xlabel("Actual brightness")
plt.ylabel("Predicted brightness")
plt.title("Predicted vs Actual brightness (NN)")
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 11. Helper: predict brightness for a new sequence
# ------------------------------------------------

def predict_brightness(seq: str) -> float:
    encoded = one_hot_encode(seq, max_len).reshape(1, -1)
    encoded_scaled = scaler.transform(encoded)
    pred = model.predict(encoded_scaled, verbose=0)[0, 0]
    return float(pred)

# Example sanity check on first sequence
example_seq = seqs[0]
print("Example name:", proteins[0].get("name"))
print("True brightness:", brightness_vals[0])
print("Predicted brightness:", predict_brightness(example_seq))
```

```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras import layers, models

# ------------------------------------------------
# 1. Download Aequorea victoria proteins (parent_organism=6100)
# ------------------------------------------------

proteins = []
url = "https://www.fpbase.org/api/proteins/?parent_organism=6100&format=json"

while url:
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    if isinstance(data, list):
        proteins.extend(data)
        break
    else:
        proteins.extend(data.get("results", []))
        url = data.get("next")

# Optional: remove families you don't want
exclude_terms = ('channelrhodopsin', 'rcamp', 'gcamp2', 'cp-mkate', 'cegfp')
proteins = [
    p for p in proteins
    if not any(term in (p.get('name') or '').lower() for term in exclude_terms)
]

print(f"Total protein records retrieved (A. victoria filtered): {len(proteins)}")

# ------------------------------------------------
# 2. Extract sequences and brightness
# ------------------------------------------------

seqs = []
brightness_vals = []

for entry in proteins:
    seq = entry.get("seq")
    states = entry.get("states", [])
    brightness = None

    if states and isinstance(states, list):
        # Try brightness, fallback to rel_brightness if needed
        for st in states:
            b = st.get("brightness")
            if b is None:
                b = st.get("rel_brightness")
            if b is not None:
                brightness = b
                break

    if seq and brightness is not None:
        seqs.append(seq)
        brightness_vals.append(brightness)

print(f"Total usable sequences with brightness: {len(seqs)}")

# Bail out early if nothing usable
if len(seqs) == 0:
    raise RuntimeError("No sequences with brightness found. Check FPbase fields.")

# ------------------------------------------------
# 3. One-hot encode sequences (fixed length = 238)
# ------------------------------------------------

max_len = 238  # EGFP length (truncate or zero-pad others)
aa_alphabet = 'ACDEFGHIKLMNPQRSTVWY'
aa_index = {aa: i for i, aa in enumerate(aa_alphabet)}

def one_hot_encode(seq: str, max_len: int) -> np.ndarray:
    arr = np.zeros((max_len, len(aa_alphabet)), dtype=np.float32)
    for i, aa in enumerate(seq[:max_len]):
        if aa in aa_index:
            arr[i, aa_index[aa]] = 1.0
    return arr.flatten()

X = np.array([one_hot_encode(seq, max_len) for seq in seqs], dtype=np.float32)
y = np.array(brightness_vals, dtype=np.float32)

print("X shape:", X.shape)  # (N, 238*20)
print("y shape:", y.shape)
print(f"Brightness range: min={y.min():.2f}, max={y.max():.2f}")

# ------------------------------------------------
# 4. Quick look at brightness distribution
# ------------------------------------------------

plt.figure(figsize=(6, 4))
plt.hist(y, bins=20)
plt.xlabel("Brightness")
plt.ylabel("Count")
plt.title("Brightness distribution (A. victoria set)")
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 5. Train/test split and scaling
# ------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)

# Standardize features (helps NN training even though it's one-hot)
scaler = StandardScaler(with_mean=True, with_std=True)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n=== Data prepared for NN training ===")
print(f"Training samples: {X_train_scaled.shape[0]}")
print(f"Test samples:     {X_test_scaled.shape[0]}")
print(f"Feature dim:      {X_train_scaled.shape[1]}")

# ------------------------------------------------
# 6. Define neural network regressor
# ------------------------------------------------

input_dim = X_train_scaled.shape[1]

model = models.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # brightness (regression)
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='mse',
    metrics=['mae']
)

model.summary()

# ------------------------------------------------
# 7. Train the model
# ------------------------------------------------

print("\n=== Starting neural network training (brightness) ===")
print("Hyperparameters: epochs=80, batch_size=32, optimizer=Adam(1e-3)\n")

history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=80,
    batch_size=32,
    verbose=1
)

print("\n=== Training complete ===")

# ------------------------------------------------
# 8. Plot training curves (loss & MAE)
# ------------------------------------------------

history_dict = history.history

plt.figure(figsize=(6, 4))
plt.plot(history_dict['loss'], label='Train loss')
plt.plot(history_dict['val_loss'], label='Val loss')
plt.xlabel("Epoch")
plt.ylabel("MSE loss")
plt.title("Training vs Validation Loss (brightness)")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(history_dict['mae'], label='Train MAE')
plt.plot(history_dict['val_mae'], label='Val MAE')
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.title("Training vs Validation MAE (brightness)")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 9. Evaluate on held-out test data
# ------------------------------------------------

test_loss, test_mae = model.evaluate(X_test_scaled, y_test, verbose=0)
print("\n=== Neural Network Evaluation (held-out test set) ===")
print(f"  Test MSE: {test_loss:.3f}")
print(f"  Test MAE: {test_mae:.3f}")

# ------------------------------------------------
# 10. Predicted vs actual brightness scatter
# ------------------------------------------------

y_pred = model.predict(X_test_scaled, verbose=0).reshape(-1)

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--')
plt.xlabel("Actual brightness")
plt.ylabel("Predicted brightness")
plt.title("Predicted vs Actual brightness (NN)")
plt.tight_layout()
plt.show()

# ------------------------------------------------
# 11. Helper: predict brightness for a new sequence
# ------------------------------------------------

def predict_brightness(seq: str) -> float:
    encoded = one_hot_encode(seq, max_len).reshape(1, -1)
    encoded_scaled = scaler.transform(encoded)
    pred = model.predict(encoded_scaled, verbose=0)[0, 0]
    return float(pred)

# Example sanity check on first sequence
example_seq = seqs[0]
print("\n=== Sanity check on first sequence ===")
print("Example name:", proteins[0].get("name"))
print("True brightness:", brightness_vals[0])
print("Predicted brightness:", predict_brightness(example_seq))

# ------------------------------------------------
# 12. Single-site mutational scan to find brightness-enhancing mutations
# ------------------------------------------------

# Choose which sequence to use as the parent for design
parent_index = 0  # change this to pick a different sequence

parent_seq = seqs[parent_index]
parent_brightness_true = brightness_vals[parent_index]
parent_brightness_pred = predict_brightness(parent_seq)

parent_name = proteins[parent_index].get("name") or f"seq_{parent_index}"

print("\n=== Mutational scan on parent sequence ===")
print(f"Parent index: {parent_index}")
print(f"Parent name:  {parent_name}")
print(f"Parent length: {len(parent_seq)} aa")
print(f"True brightness (FPbase): {parent_brightness_true:.3f}")
print(f"Predicted brightness (NN): {parent_brightness_pred:.3f}\n")

mut_results = []  # (delta, mut_pred, pos, wt, aa, mutant_seq)

# Determine valid positions (only standard AAs)
valid_positions = [i for i, aa in enumerate(parent_seq) if aa in aa_alphabet]
total_mutants = len(valid_positions) * (len(aa_alphabet) - 1)

print(f"Scanning single-site mutants for parent {parent_name} ...")
print(f"  Valid positions:        {len(valid_positions)}")
print(f"  Mutants per position:   {len(aa_alphabet) - 1}")
print(f"  Total mutants to score: {total_mutants}")
print("  (Progress will be printed every 500 mutants)\n")

mut_count = 0
for pos in valid_positions:
    wt = parent_seq[pos]

    for aa in aa_alphabet:
        if aa == wt:
            continue  # skip wild-type

        mutant_seq = parent_seq[:pos] + aa + parent_seq[pos+1:]
        mut_pred = predict_brightness(mutant_seq)
        delta = mut_pred - parent_brightness_pred

        mut_results.append((delta, mut_pred, pos, wt, aa, mutant_seq))

        mut_count += 1
        if mut_count % 500 == 0 or mut_count == total_mutants:
            pct = 100.0 * mut_count / total_mutants
            print(f"  Scored {mut_count}/{total_mutants} mutants "
                  f"({pct:5.1f}%)")

print("\nFinished mutational scan.")
print(f"Total mutants scored: {len(mut_results)}\n")

# Sort mutants by improvement (largest Δ first)
mut_results.sort(reverse=True, key=lambda x: x[0])

# Print the top N mutants
top_n = 20
print(f"Top {top_n} single mutants by Δ(predicted brightness):")
for i, (delta, mut_pred, pos, wt, aa, mutant_seq) in enumerate(mut_results[:top_n], start=1):
    print(
        f"{i:2d}. Pos {pos+1:3d} {wt}->{aa}  "
        f"Δpred = {delta:+.3f}  pred = {mut_pred:.3f}"
    )

# Optional: best mutation per position
best_by_position = {}
for delta, mut_pred, pos, wt, aa, mutant_seq in mut_results:
    if pos not in best_by_position:
        best_by_position[pos] = (delta, mut_pred, wt, aa)

print("\nBest mutation per position (first 20 positions shown):")
for pos in sorted(best_by_position.keys())[:20]:
    delta, mut_pred, wt, aa = best_by_position[pos]
    print(
        f"Pos {pos+1:3d}: {wt}->{aa}  "
        f"{wt}->{aa}  Δpred = {delta:+.3f}  pred = {mut_pred:.3f}"
    )

# ------------------------------------------------
# 13. Visualize single-site mutational scan
# ------------------------------------------------

# Turn mutational results into a DataFrame
mut_df = pd.DataFrame(
    mut_results,
    columns=["delta", "pred", "pos0", "wt", "aa", "seq"]
)
mut_df["pos"] = mut_df["pos0"] + 1  # 1-based positions for labeling
mut_df["label"] = mut_df["wt"] + mut_df["pos"].astype(str) + mut_df["aa"]

# ----- Summary printouts -----
print("\n=== Mutational scan summary ===")
print(f"Parent predicted brightness: {parent_brightness_pred:.3f}")
print(f"Total single mutants scored: {len(mut_df)}")

mean_pred = mut_df["pred"].mean()
std_pred = mut_df["pred"].std()
mean_delta = mut_df["delta"].mean()
num_better = (mut_df["delta"] > 0).sum()
num_worse = (mut_df["delta"] < 0).sum()

print(f"Mean predicted brightness (mutants): {mean_pred:.3f}")
print(f"Std predicted brightness (mutants):  {std_pred:.3f}")
print(f"Mean Δpred (mutant - parent):        {mean_delta:+.3f}")
print(f"Number improved (Δpred > 0):          {num_better}")
print(f"Number worse    (Δpred < 0):          {num_worse}")

# Top 10 mutants
top10 = mut_df.sort_values("delta", ascending=False).head(10)
print("\nTop 10 predicted brightness-enhancing single mutants:")
for _, row in top10.iterrows():
    print(
        f"  {row['label']:>7}  pos={int(row['pos']):3d}  "
        f"pred={row['pred']:.3f}  Δpred={row['delta']:+.3f}"
    )

# ----- Scatter plot like your figure -----
upper = mean_pred + 2 * std_pred
lower = mean_pred - 2 * std_pred

plt.figure(figsize=(7, 5))
plt.scatter(mut_df["delta"], mut_df["pred"], s=15, color="black", alpha=0.7)

# Mean and ±2σ lines (brightness axis)
plt.axhline(mean_pred, linestyle="--", color="green",
            label=f"Mean = {mean_pred:.2f}")
plt.axhline(upper, linestyle=":", color="red",
            label=f"+2σ = {upper:.2f}")
plt.axhline(lower, linestyle=":", color="blue",
            label=f"-2σ = {lower:.2f}")

plt.xlabel("Δ Predicted Brightness (mutant - parent)")
plt.ylabel("Predicted Brightness")
plt.title("Predicted Brightness of Single Point Mutants (±2 SD Outliers)")

# Label outliers (beyond ±2 SD)
outliers = mut_df[(mut_df["pred"] >= upper) | (mut_df["pred"] <= lower)]
for _, row in outliers.iterrows():
    y_offset = 0.2 if row["pred"] >= upper else -0.4
    color = "red" if row["pred"] >= upper else "blue"
    plt.text(
        row["delta"], row["pred"] + y_offset,
        row["label"],
        fontsize=6, ha="center", va="bottom", color=color
    )

plt.legend()
plt.tight_layout()
plt.show()

# ----- Heatmap of Δpred (AA vs position) -----
# Pivot: rows = AA, columns = position, values = Δpred
heat_df = mut_df.pivot_table(
    index="aa", columns="pos", values="delta", aggfunc="mean"
)

# Optionally restrict to first N positions to keep the figure readable
max_positions_to_plot = 290  # change if you want more/less
cols_sorted = sorted(heat_df.columns)[:max_positions_to_plot]
heat_sub = heat_df[cols_sorted]

vmax = np.nanmax(np.abs(heat_sub.values))  # symmetric color scale

plt.figure(figsize=(10, 4))
im = plt.imshow(
    heat_sub.values,
    aspect="auto",
    cmap="coolwarm",
    vmin=-vmax,
    vmax=vmax
)
plt.colorbar(im, label="Δ Predicted Brightness (mutant - parent)")
plt.yticks(range(len(heat_sub.index)), heat_sub.index)
plt.xticks(
    range(len(heat_sub.columns)),
    heat_sub.columns,
    rotation=90
)
plt.xlabel("Position (1-based)")
plt.ylabel("Mutant AA")
plt.title("Single-Mutant ΔPredicted Brightness Heatmap")
plt.tight_layout()
plt.show()

print("\n=== All analysis complete ===")
print("Figures generated:")
print("  - Brightness histogram")
print("  - Training vs validation loss & MAE")
print("  - Predicted vs actual brightness scatter")
print("  - Mutational scan scatter (±2 SD outliers)")
print("  - Mutational Δbrightness heatmap")
```
